```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('CKD.csv')
```

```python
data
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data
```

```python
data.columns
```

```python
x=data[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes' ]]
```

```python
x
```

```python
x.shape
```

```python
y=data[['classification_yes']]
```

```python
y
```

```python
data['classification_yes'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test =sc.transform(x_test)
```

```python
from sklearn.svm import SVC


```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'kernel':['linear','rbf','sigmoid','poly'],
            'gamma':['auto','scale'],
            'C':[10,100,1000,2000,3000]}
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
y_pred=grid.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)
```

```python

clf
```

```python
cm
```

```python

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```
